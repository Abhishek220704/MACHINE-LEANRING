‚úÖ 1. Why Linear Regression Fails for Classification?

Linear Regression outputs continuous values like:

3.2, 7.8, 20.5‚Ä¶


But classification needs 0 or 1
(Yes/No, Spam/Not Spam, Buy/Not Buy)

So‚Ä¶

We need a function that converts numbers into probability (between 0 and 1)

üëâ Sigmoid Function

‚≠ê 2. Sigmoid Function

This is the heart of logistic regression.

œÉ(x) = 1 / (1 + e^(-x))


It converts any value into a probability.

Example:

x = 2 ‚Üí œÉ(2) ‚âà 0.88
x = -3 ‚Üí œÉ(-3) ‚âà 0.04

‚≠ê 3. Logistic Regression Formula

Prediction:

y_pred = sigmoid(w*x + b)


If y_pred > 0.5 ‚Üí Class = 1
Else ‚Üí Class = 0

‚≠ê 4. Loss Function: Binary Cross Entropy

This measures how wrong our predictions are.

Loss = -( y*log(y_pred) + (1-y)*log(1-y_pred) )


Gradient descent is used to update:

w_new = w_old - lr * derivative
b_new = b_old - lr * derivative

**Linear Regression**

```python
import numpy as np

# Sample dataset: hours studied vs pass/fail
X = np.array([1,2,3,4,5], dtype=float)
Y = np.array([0,0,0,1,1], dtype=float)

# reshape for matrix operations
X = X.reshape(-1,1)

# initialize parameters
w = 0.1
b = 0.1
lr = 0.1

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# training
for epoch in range(1000):
    z = w*X + b
    y_pred = sigmoid(z)

    dw = np.mean((y_pred - Y) * X)
    db = np.mean(y_pred - Y)

    w = w - lr * dw
    b = b - lr * db

print("Weight:", w)
print("Bias:", b)
print("Prediction for 4 hours:", sigmoid(w*4 + b))
```

```python
from sklearn.linear_model import LogisticRegression
import numpy as np

X = np.array([[1],[2],[3],[4],[5]])
Y = np.array([0,0,0,1,1])

model = LogisticRegression()
model.fit(X, Y)

print("Prediction for 4 hours:", model.predict([[4]]))
print("Probability:", model.predict_proba([[4]]))
```
