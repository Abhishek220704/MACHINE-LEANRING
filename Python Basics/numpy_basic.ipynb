# **NumPy Basics**

```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print("a + b =", a + b)
print("Dot product =", np.dot(a, b))
print("Matrix Multiplication =", a @ b)
```
