What is Probability? (Simple ML Explanation)

Probability tells us how likely something is to happen.

Examples in ML:

Probability an email is spam

Probability a customer will buy

Probability a pixel belongs to a cat/dog

Formula:

P(Event) = (Number of favorable outcomes) / (Total outcomes)


Example:

P(rolling a 3 on dice) = 1/6

✅ 2. Types of Probability

1️⃣ Prior Probability

Your belief before seeing data.
Used in Naive Bayes.

Example:

P(Spam) = 0.2  (20% emails are spam)

2️⃣ Conditional Probability

Probability of A happening given B happened.

P(A | B)


Example:

Probability an email is spam given it contains "win money".

3️⃣ Joint Probability

Probability of A and B happening together.

P(A ∩ B)


Used heavily in Naive Bayes.

⭐ 3. Bayes Theorem — VERY Important for ML

This is the formula behind Naive Bayes classifier.

P(A|B) = [ P(B|A) * P(A) ] / P(B)


Meaning:
Probability of A happening given B.

Simple example in ML:

Probability email is spam given it contains "free".

⭐ 4. Statistics Needed for ML
1️⃣ Mean (Average)
import numpy as np
data = [1, 2, 3, 4, 5]
np.mean(data)

2️⃣ Median

Middle value.

np.median(data)

3️⃣ Variance

How spread out the data is.

np.var(data)

4️⃣ Standard Deviation

Square root of variance.

np.std(data)


Used to detect outliers, normalization, etc.

5️⃣ Normal Distribution (Bell Curve)

Most natural data follows this.

ML uses it in:

Gaussian models

Naive Bayes

Probability density estimation

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(50, 10, 1000)  # mean=50, sd=10

plt.hist(data, bins=30)
plt.title("Normal Distribution")
plt.show()
```
