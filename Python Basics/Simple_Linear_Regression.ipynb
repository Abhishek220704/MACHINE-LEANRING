üëâ Model learns the pattern

Examples:

Predict house price ‚Üí Regression

Predict spam/not spam ‚Üí Classification

Predict marks from study hours ‚Üí Regression

‚≠ê 2. What is Linear Regression?

Linear regression tries to learn a line:

y = m*x + c


Where:

x = input

y = output

m = slope (weight)

c = intercept (bias)

Goal:
üëâ Choose m and c such that prediction error is minimum.

‚≠ê 3. Loss Function (Very Important)

To measure error, we use Mean Squared Error (MSE):

MSE = average( (y_pred ‚Äì y_actual)^2 )

‚≠ê 4. How the Model Learns?

Using Gradient Descent (you learned it earlier):

w_new = w_old - learning_rate * derivative


**Linear Regression**

```python
import numpy as np
import matplotlib.pyplot as plt

# Dataset: study hours vs marks
X = np.array([1,2,3,4,5], dtype=float)
Y = np.array([30, 40, 50, 60, 70], dtype=float)

m = 0  # slope
c = 0  # intercept
lr = 0.01  # learning rate
epochs = 10

n = len(X)

for i in range(epochs):
    Y_pred = m*X + c

    # compute gradients
    dm = (-2/n) * sum(X * (Y - Y_pred))
    dc = (-2/n) * sum(Y - Y_pred)

    # update parameters
    m = m - lr*dm
    c = c - lr*dc

print("Slope (m):", m)
print("Intercept (c):", c)

# Plot final line
plt.scatter(X, Y)
plt.plot(X, m*X + c, color='red')
plt.title("Linear Regression")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

X = np.array([[1],[2],[3],[4],[5]])
Y = np.array([30,40,50,60,70])

model = LinearRegression()
model.fit(X, Y)

print("Slope:", model.coef_)
print("Intercept:", model.intercept_)

print("Prediction for 6 hours:", model.predict([[6]]))
```
